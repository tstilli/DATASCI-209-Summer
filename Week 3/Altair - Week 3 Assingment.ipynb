```python
alt.Chart(iris).mark_point().encode(
    x='sepalLength',
    y='sepalWidth',
    color = 'species',
    tooltip=['sepalLength', 'sepalWidth', 'petalLength', 'petalWidth']
).interactive().properties(
    width=800,
    height=500
)
```
