```python
import pandas as pd
```

```python
data = pd.read_csv('players_20.csv', dtype = 'object')
data.head(1)
```

```python
data['age'] = data['age'].astype(float)
data['sofifa_id'] = data['sofifa_id'].astype(str)
```

```python
table = data.groupby(['nationality']).agg(mean = ('age', 'mean'),count = ('sofifa_id', 'size'))
table = table.rename(columns={'mean':'Average Age','count':'Count of Players'})
table.head(10)
```
